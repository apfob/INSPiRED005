```python
from radiomics import featureextractor
```

```python
extractor = featureextractor.RadiomicsFeatureExtractor(resampledPixelSpacing = [1,1,1], interpolator = 3, binWidth = 10, 
                                                      verbose = True, resegmentMode = [ 'sigma',[-3,3]])
# If we set a fixed bincount at here, there will have a warning that they recommend a fixed binWidth only
```

```python
list_1 = ["{0:03}".format(i) for i in range(1,600)]
list_2 = ["{0:02}".format(f) for f in range(13)]
```

```python
data_STR_clahe = pd.DataFrame()
```

```python
for i in list_1:      
    for f in list_2:
        imgname = "STR1_VTIQ"+ str(f) + str(i) + "_hist_clahe.nrrd"
        if os.path.exists(imgname):
            img = sitk.ReadImage(imgname, sitk.sitkInt8)
            maskname = "STR1_VTIQ"+ str(f) + str(i) + "_hist_clahe_TPS.nrrd"
            if os.path.exists(maskname):
                mask = sitk.ReadImage(maskname, sitk.sitkInt8)
                try:
                    featureVector2 = extractor.execute(img, mask)
                    featureVector2["Name"]=maskname
                    df_add = pd.DataFrame([featureVector2])
                    data_STR_clahe = pd.concat([data_STR_clahe,df_add])
                except ValueError:
                    print(maskname) 
            else:
                continue
```
